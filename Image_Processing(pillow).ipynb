```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
urls=["http://iiif.ncsu.edu/iiif/005274/full/800,/0/default.jpg",
      "http://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
      "http://placekitten.com/800/571"]
for url in urls:
 image=io.imread(url)
 image_2=cv.cvtColor(image,cv.COLOR_BGR2RGB)
 final_frame=cv.cvtColor((image,image_2))
 cv2_imshow(final_frame)
print('\n')
```
